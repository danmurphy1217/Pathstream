```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
raw_df = pd.read_csv("avocado.csv")
raw_df.columns = raw_df.columns.str.replace(' ', '_')
raw_df.head(10)
```

```python
volume_df = raw_df['Total_Volume'].apply(lambda x : math.log10(x))
volume_df.head(10)
```

```python
print("Min Value: " + str(np.quantile(volume_df, 0)))
print("Q1: " + str(np.quantile(volume_df, .25)))
print("Median: " + str(np.quantile(volume_df, .5)))
print("Q3: " + str(np.quantile(volume_df, .75)))
print("Max Value: " + str(np.quantile(volume_df, 1)))
```

```python
upper_volume = raw_df[raw_df['Total_Volume'].apply(lambda x : math.log10(x)) > np.quantile(volume_df, .5)]
lower_volume = raw_df[raw_df['Total_Volume'].apply(lambda x: math.log10(x)) < np.quantile(volume_df, .5)]
```

```python
[pd.value_counts(upper_volume[colName]) for colName in raw_df.columns]
```

```python
[pd.value_counts(lower_volume[colName]) for colName in raw_df.columns]
```

```python
log_lower_volume = lower_volume['Total_Volume'].apply(lambda x: math.log10(int(x)))
log_upper_volume = upper_volume['Total_Volume'].apply(lambda x: math.log10(x))


np.sort(log_lower_volume)
```

```python
np.sort(log_upper_volume)
```

```python
cali = pd.DataFrame(raw_df[raw_df['region'] == 'California'])
cali.sort_values(by = 'Date')
vol = cali['Total_Volume'].apply(lambda x : math.log10(x))
```

```python
sns.scatterplot(x = cali['Date'], y = vol, data = cali, hue='type')
```

# Solved... it is conventional vs. organic. The volume for organic is much lower (interesting!)
